**8.Write the code to find the percentage missing value (column-wise) in master_dataframe. Drop  unnecessary columns.**


```python
missing_percentage = master_dataframe.isnull().mean() * 100
missing_percentage
```

```python
#removing columns that have more than 50 % of data is missing
columns_to_drop = missing_percentage[missing_percentage > 50].index
master_dataframe = master_dataframe.drop(columns=columns_to_drop)
```

**11.Identify the top 9 countries in terms of highest investment in ‘venture’ fund type.**

```python
# Filtering
venture_funding_df = master_dataframe[master_dataframe['funding_round_type'] == 'venture']

# Grouping by country and sum
country_investment = venture_funding_df.groupby('country_code')['raised_amount_usd'].sum()

# sorting to get top 9
top_9_countries = country_investment.sort_values(ascending=False).head(9)
```

```python
top_9_countries
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_9_countries.index, y=top_9_countries.values, palette='viridis')
plt.title('Top 9 Countries by Investment in Venture Funding')
plt.xlabel('Country')
plt.ylabel('Total Investment (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
```

**13.Using the user defined function convert the ‘rasied_amount_used’ column into a categorical  column as follow-**

**a)If the amount is less than 5 Millions then recode as TypeA.**

**b)If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.**

**c)If the amount is greater than 7 Millions then recode as TypeC.**


```python
def categorize_amount(amount):
    if amount < 5000000:
        return 'TypeA'
    elif 5000000 <= amount < 7000000:
        return 'TypeB'
    else:
        return 'TypeC'
```

```python
#applying the function to the raised_amount_usd column
master_dataframe['amount_category'] = master_dataframe['raised_amount_usd'].apply(categorize_amount)
```
